```python
import pandas as pd
from random import randint
import random
import scipy.stats
random.seed(42)
```

```python
param = [randint(1, 10) for _ in range(100)]
users = ['Пользователь' for _ in range(100)]
# Сброс ограничений на количество выводимых рядов
pd.set_option('display.max_rows', None)
 
# Сброс ограничений на число столбцов
pd.set_option('display.max_columns', None)
 
# Сброс ограничений на количество символов в записи
pd.set_option('display.max_colwidth', None)
```

```python
data = pd.DataFrame({'Наименование' : users,
                     'Параметр(кол-во пройденных уроков)' : param})
data
```

```python
grouped_data = data.groupby('Параметр(кол-во пройденных уроков)', as_index=False).agg({"Наименование" : 'count'})
grouped_data
```

```python
grouped_data_with_freqs = grouped_data['Наименование'] / 100
grouped_data_with_freqs
```

```python
Mx = sum(grouped_data['Наименование'] * grouped_data['Параметр(кол-во пройденных уроков)'])
```

```python
grouped_data.var()
```
