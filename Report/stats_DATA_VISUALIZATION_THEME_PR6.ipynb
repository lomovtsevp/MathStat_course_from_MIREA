#***2. Обзор различных видов визуализации данных***#

***2.1 ScatterPlot (Диаграмма рассеяния)***

Точечная диаграмма (также называемая диаграммой рассеяния, диаграммой рассеяния, диаграммой рассеяния, диаграммой рассеяния или диаграммой рассеяния) - это тип диаграммы или математической диаграммы, использующей декартовы координаты для отображения значений обычно двух переменных для набора данных.

```python
plt.figure(figsize=(6, 6))
plt.title('Диаграмма рассеяния')
plt.xlabel('Средний доход домашних хозяйств')
plt.ylabel('Средняя стоимость жилья')
plt.scatter(data.median_income, data.median_house_value, color='blue', s=25)
plt.show()
```

***Замечание.*** Диаграмма рассеяния помогает увидеть взаимосвязь между двумя количественными переменными. Поэтому на этапе предобработки и первичного анализа данных чаще всего используют именно данный вид графика.

***2.2 Histogram (Гистограмма частот)***

Гистограмма — способ представления табличных данных в графическом виде — в виде столбчатой диаграммы.

Количественные соотношения некоторого показателя представлены в виде прямоугольников, площади которых пропорциональны. Чаще всего для удобства восприятия ширину прямоугольников берут одинаковую, при этом их высота определяет соотношения отображаемого параметра.

```python
plt.figure(figsize=(6, 6))
plt.xlabel('Общее количество жильцов в районе')
plt.ylabel('Частота')
sns.histplot(data.households, color='green').set_title('Гистограмма частот');
```

***Замечание.*** Гистограмму частот так же строят в процессе первичного анализа данных для оценки распределения категориальной переменной.

***2.3 Boxplot (Ящик с усами)***

Ящик с усами или диаграмма размаха — график, использующийся в описательной статистике, компактно изображающий одномерное распределение вероятностей.

Такой вид диаграммы в удобной форме показывает медиану (или, если нужно, среднее), нижний и верхний квартили, минимальное и максимальное значение выборки и выбросы. Несколько таких ящиков можно нарисовать бок о бок, чтобы визуально сравнивать одно распределение с другим; их можно располагать как горизонтально, так и вертикально. Расстояния между различными частями ящика позволяют определить степень разброса (дисперсии) и асимметрии данных и выявить выбросы.

```python
plt.title('Ящик с усами (boxplot)')
sns.boxplot(x=data.median_house_value, color='orange');
```

***Пояснения к графику.*** Для правильной интерпретации данного вида визуализации данных необходимо определить, что он показывает. Сам ящик показывает 25 процентиль (1 квартиль), 50 процентиль(медиану или 2 квартиль), 75 процентиль (3 квартиль) в виде вертикальных линий - границ ящика и линии посередине. Усы показывают $1,5 * IQR$, где <p>$IQR$ = $Quartile_3 - Quartile_1, IQR$ - интерквартильный размах. <p>Все объекты, которые не вошли в этот диапазон, принято считать выбросами.

Таким образом ***Boxplot*** позволяет идентифицировать выбросы и оценивать разброс данных.

***2.4 Barplot (Столбчатая диаграмма)***

Столбчатая диаграмма — диаграмма, представленная прямоугольными зонами (столбцами), высоты или длины которых пропорциональны величинам, которые они отображают.

Столбчатую диаграмму легко спутать с гистограммой. Отличие в том, что в стобчатой диаграмме участвуют две переменных по осям, а в гистограмме - одна.



```python
plt.title('Столбчатая диаграмма')

fih = sns.barplot(data.latitude[:5], data.households[:5]);
fih.set(xlabel='Географическая широта', ylabel='Общее количество жильцов в районе');
```

***Замечание***. Таким образом столбчатая диаграмма показывает зависимость одной переменной от другой. Такие графики, к сожалению, на больших данных достаточно неинформативны.

***2.5 HeatMap (Тепловая карта)***

Тепловая карта (англ. heatmap) — графическое представление данных, где индивидуальные значения в таблице отображаются при помощи цвета. Термин «heatmap» изначально был придуман и официально зарегистрирован как товарный знак разработчиком программного обеспечения Кормаком Кинни.

Тепловую карту очень часто применяют для ***корреляционной матрицы признаков исходных данных***. Ниже приведен пример такой тепловой карты

```python
plt.title('Тепловая карта коэффициентов корреляции признаков')
sns.heatmap(data.corr());
```

***Как интерпретировать результаты визуализации данных методом тепловой карты?***

Справа от графика нарисована шкала коэффициентов корреляции Спирмена и соответствующих им оттенкам цветов. На осях расположены признаки. На пересечении *i-го* и *j-го* признаков обозначена сила линейной взаимосвязи между признаками *i* и *j*.

Таким образом можно определять, насколько сильно коррелируют признаки в данных.

***2.6 Violin Plot (Скрипичная диаграмма)***

График скрипки - это метод визуализации числовых данных. Он похож на  диаграмму разброса ***(boxplot)*** с добавлением повернутого графика ядерной оценки плотности с каждой стороны.

График сам по себе довольно эстетичен, однако на практике редко применяют данный тип визуализации вследствие низкой информативности.

```python
fig = sns.violinplot(data.households, color='violet')
fig.set(xlabel='Общее количество жильцов');
```

Таким образом график скрипки (***violin plot***) показывает ядерную оценку плотности распределения величины, но достаточно трудно интерпретируем - лучше воспользоваться другими методами.

***2.7 Pair Plot***

Данный вид графиков, на мой взгляд, наиболее информативен, так как он показывает всевозможные комбинации признаков и их диаграммы рассеяния, а на главной диагонали строится гистограмма признака. На большом признаковом пространстве данный метод будет работать достаточно долго.

```python
sns.pairplot(data);
```

Таким образом Pair Plot довольно эффективно отображает картину данных.

***2.8 Regression Plot***

Данный график автоматически строит линейный тренд между двумя признаками, показывая доверительный интервал. Позволяет предсказывать значения.



```python
plt.title('Линейная регрессия')
sns.regplot(data.median_income, data.median_house_value, color='red', marker='*');
```

Данный график позволяет быстро построить линейный тренд и предсказать значение целевой переменной.

***2.9. RidgeLine Plot (Горбатый график)***

Горбатый график - это набор перекрывающихся графиков плотности, которые помогают сравнивать несколько распределений между наборами данных. Участки хребтовой линии выглядят как горный хребет, они могут быть весьма полезны для визуализации изменений в распределении во времени или пространстве. Довольно хорошо показывает распределения данных в разрезе категориальной переменной

```python
rp = sns.FacetGrid(data, row="commerce", hue="commerce", aspect=5, height=1.25, size=2);
rp.map(sns.kdeplot, 'median_house_value', clip_on=False, shade=True, alpha=0.7, lw=4, bw=.2);
rp.map(plt.axhline, y=0, lw=4, clip_on=False);
```

***Замечание.*** Горбатый график достаточно хорошо показывает распределение количественной переменной при разбиении выборки по категориальному признаку
