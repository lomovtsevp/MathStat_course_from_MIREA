```python
rp = sns.FacetGrid(data, row="commerce", hue="commerce", aspect=5, height=1.25, size=2);
rp.map(sns.kdeplot, 'median_house_value', clip_on=False, shade=True, alpha=0.7, lw=4, bw=.2);
rp.map(plt.axhline, y=0, lw=4, clip_on=False);
```
